```python
import pandas as pd
df_debt = pd.read_csv("IDSData.csv")
df_debt = df_debt.drop(df_debt.loc[:, "2019":"Unnamed: 61"].columns, axis = 1) 
df_debt = df_debt[["Country Name", "2018"]]
df_debt = df_debt.rename(columns={"2018" : "Debt"})
df_debt = df_debt.groupby("Country Name")["Debt"].sum()
df_debt = pd.DataFrame({"Country Name" : list(df_debt.index),
             "Debt" : df_debt.values})
df_debt.head()
```

```python
df_GDP = pd.read_csv("GDP.csv", skiprows=4, header=None)
df_GDP.columns = df_GDP.iloc[0]
df_GDP.columns = df_GDP.columns.astype(str).str.replace("\.0", "")
df_GDP = df_GDP[["Country Name", "2019"]]
df_GDP.rename(columns={"2019": "GDP"}, inplace=True)
df_GDP.head()
```

```python
df_Population = pd.read_csv("Population.csv", skiprows=3, header=None)
df_Population.columns = df_Population.iloc[0]
df_Population.columns = df_Population.columns.astype(str).str.replace("\.0", "")
df_Population = df_Population[["Country Name", "2019"]]
df_Population.rename(columns={"2019": "Population"}, inplace=True)
df_Population.head()
```

```python
df = df_GDP.merge(df_Population, on="Country Name", how="outer").merge(df_debt, on="Country Name", how="outer")
df.to_csv("df.csv", index=False)
!gopen df.csv
```
