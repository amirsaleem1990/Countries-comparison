```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Original_data_converted_to_csv_from_rds_file.csv")
num_vars = data.dtypes[(data.dtypes == "int64") | (data.dtypes == "float64")].index.drop(["Unnamed: 0",
                                                                                         "30 DPDs Instance",
                                                                                         "applicationnumber"])
old = data[data.DataType == "Old"].drop(["Unnamed: 0", "DataType", "30 DPDs Instance"], axis=1)
new = data[data.DataType == 'New'].drop(["Unnamed: 0", "DataType", "30 DPDs Instance"], axis=1)
```

```python
names = []
importanc = []
rnd_clf = RandomForestClassifier(n_estimators=500, n_jobs=50)
rnd_clf.fit(old[num_vars].drop("DPD30", axis=1), old.DPD30)
for name, importance in zip(old[num_vars].drop("DPD30", axis=1).columns, rnd_clf.feature_importances_):
    names.append(name)
    importanc.append(importance)
df = pd.DataFrame()
for i in ['names', 'importanc']:
    df[i] = eval(i)
# df
```

```python
plt.barh(df.names, df.importanc)
```

```python
rnd_clf2 = RandomForestClassifier(n_estimators=500, n_jobs=40)
X_new = new[num_vars].drop(["AccomodationType","ResidenceType"], axis=1)
rnd_clf2.fit(X_new.drop("DPD30", axis=1), new.DPD30)


names = []
importanc = []
for name, importance in zip(new[num_vars].drop("DPD30", axis=1).columns, rnd_clf2.feature_importances_):
    names.append(name)
    importanc.append(importance)

df2 = pd.DataFrame()
for i in ['names', 'importanc']:
    df2[i] = eval(i)
plt.barh(df2.names, df2.importanc)
```

```python
df3 = pd.read_csv("in_both_new_and_old.csv")
df3.DataType.value_counts()
```

```python
old = df3[df3.DataType == "Old"].drop(["Unnamed: 0", "DataType", "30 DPDs Instance"], axis=1)
new = df3[df3.DataType == "New"].drop(["Unnamed: 0", "DataType", "30 DPDs Instance"], axis=1)
```

```python
rnd_clf3 = RandomForestClassifier(n_estimators=500, n_jobs=40)
X3_new = new[num_vars].drop(["AccomodationType","ResidenceType"], axis=1)
rnd_clf3.fit(X3_new.drop("DPD30", axis=1), new.DPD30)
names = []
importanc = []
for name, importance in zip(X3_new.drop("DPD30", axis=1).columns, rnd_clf3.feature_importances_):
    names.append(name)
    importanc.append(importance)

df3 = pd.DataFrame()
for i in ['names', 'importanc']:
    df3[i] = eval(i)
plt.barh(df3.names, df3.importanc)
```

```python
names = []
importanc = []
rnd_clf4 = RandomForestClassifier(n_estimators=500, n_jobs=50)
rnd_clf4.fit(old[num_vars].drop("DPD30", axis=1), old.DPD30)
for name, importance in zip(old[num_vars].drop("DPD30", axis=1).columns, rnd_clf4.feature_importances_):
    names.append(name)
    importanc.append(importance)
df4 = pd.DataFrame()
for i in ['names', 'importanc']:
    df4[i] = eval(i)
plt.barh(df4.names, df4.importanc)
```

```python
all_dfs = df4.merge(df3.merge(df2.merge(df, on='names'), on='names'), on="names")
all_dfs.columns = ['Name', 'Balanced_old', 'Balanced_new', 'All_new', "All_old"]
all_dfs.plot.bar(x='Name', figsize=(15.,5), title = "Variable Importanc (New VS Old)", fontsize=15, rot=90)
```
