# clustring

```python
def to_clusters(train, n=5):
    c = 0
    X = train.drop(["DPD30"], axis=1)
    c += 1
    if c == 1:
        print(X.columns)
#     n = int(input("How many clusters you want?"))
    model = KMeans(n_clusters=n, n_init=5)
    model.fit(X)
#     predictions = model.predict(test)
    train['labels'] = model.labels_
    
    all_clusturs = []
    
    for i in train.labels.unique():
        all_clusturs.append(train[train.labels == i])
    
    return all_clusturs
```

```python
def distribution_within_each_cluster(list_of_clusters):
    
    a_0 = []
    a_1 = []
    df = pd.DataFrame()
    
    for cluster in list_of_clusters:
        cluster_y_distribution = cluster.DPD30.value_counts()
        if 0 in cluster_y_distribution.index:
            a_0.append(cluster_y_distribution.loc[0])
        else:
            a_0.append(0)

        if 1 in cluster_y_distribution.index:
            a_1.append(cluster_y_distribution.loc[1])
        else:
            a_1.append(0)
    df['Size'] = np.array(a_1) + np.array(a_0)
    for i in ['a_1', 'a_0']: 
        df[i] = eval(i)
    
    zz = pd.Series({'a_1' : df['a_1'].sum(),
                   'a_0' : df['a_0'].sum(),
                   'Size': df['Size'].sum()},
                   name='AVERAGE or SUM')
    
    avg = pd.DataFrame(df.append(zz))
    avg.columns = ['Size', '1_Count', '0_Count']
    avg['1%'] = avg['1_Count'] / (avg['1_Count'] + avg['0_Count'])
    avg['0%'] = avg['0_Count'] / (avg['1_Count'] + avg['0_Count'])
    avg['Size'] = avg['1_Count'] +avg['0_Count']
    return avg
distribution_within_each_cluster(to_clusters(train))
```

# train each cluster using logistic-regression, and save results

```python
from sklearn.linear_model import LogisticRegression
from nltk import FreqDist
import os
TN = []
FP = []
FN = []
TP = []
size = []
proportion_of_1 = []
proportion_of_0 = []
Accuracy = []
for cluster in to_clusters(train):
    if len(cluster) > 50:
        X = cluster.drop(["DPD30", "labels"], axis=1)
        y = cluster.DPD30
        clf = LogisticRegression().fit(X, y)
        predicted = clf.predict(test[X.columns])
        cm = confusion_matrix(test.DPD30, predicted).ravel()
        tn, fp, fn, tp = cm / sum(cm)
        TN.append(round(tn, 3))
        FP.append(round(fp, 3))
        FN.append(round(fn, 3))
        TP.append(round(tp, 3))
        size.append(len(cluster))
        Accuracy.append(round(np.mean(predicted == test.DPD30), 3))

        t = y.value_counts()
        if 1 in t:
            proportion_of_1.append(round(dict(t)[1] / t.sum(), 2))
        else:
            proportion_of_1.append(0)
        if 0 in t:
            proportion_of_0.append(round(dict(t)[0] / t.sum(), 2))
        else:
            proportion_of_0.append(0)
d = pd.DataFrame()
for i in ['Accuracy', 'TN' ,'FP' ,'FN' ,'TP' ,'size' ,'proportion_of_1' ,'proportion_of_0']:
    d[i] = eval(i)
#     precision = [] # tp / (tp + fp)
    # recall = [] # tp / (tp + fn)
d['Precision'] = d.TP / (d.TP + d.FP)
d['Recall'] = d.TP / (d.TP + d.FN)
d = d.sort_values(by=['Precision'], ascending=False)
d
```

# our data is imbalance, so we undersample our data

```python
old = data[data.DataType == "Old"]
old.DPD30.value_counts()

zero = old[old.DPD30 == 0]
one = old[old.DPD30 == 1]
# print("1:\t", len(one))
# print("0:\t", len(zero))


# Ones
ind_list = [i for i in range(len(one))]
shuffle(ind_list)
one  = one.iloc[ind_list]
train_one = one[:round(len(one)*.8)]
test_one = one[round(len(one)*.8):]

# Zeros
ind_list = [i for i in range(len(zero))]
shuffle(ind_list)
zero  = zero.iloc[ind_list]
train_zero = zero[:round(len(zero) * (len(train_one) / len(zero)))]
test_zero = zero[round(len(zero) * (len(train_one) / len(zero))):]

# training and testing sets
train = train_one.append(train_zero)
test = test_one.append(test_zero)

print("Training Data", len(train))
print("Test Data", len(test))
```
