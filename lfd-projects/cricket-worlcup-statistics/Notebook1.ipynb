# Records / ICC Cricket World Cup, 2019

```python
import requests
from bs4 import BeautifulSoup
import os
import pandas as pd
```

# Batting averages

```python
url = "http://stats.espncricinfo.com/ci/engine/records/averages/batting.html?id=12357;type=tournament"
soup = BeautifulSoup(requests.get(url).text, "lxml")
s = soup.select("table", {"class" : "engineTable"})[0].find_all("tbody")[0].find_all("tr", {"class" : "data1"})
columns = soup.select("table", {"class" : "engineTable"})[0].select("tr", {"class" : "head"})[0].get_text().splitlines()[1:]
all_data  = []
for i in s:
    all_data.append(i.get_text().splitlines()[1:])
    df_batting_avg = pd.DataFrame(all_data, columns = columns)
    
df_batting_avg["Country"] = df_batting_avg.Player.str.extract("\(([^)]+)")
df_batting_avg["Player"] = df_batting_avg.Player.str.replace(r"\(.*\)","").str.strip().str.lower()
df_batting_avg["HS"] = df_batting_avg.HS.str.replace("*", "")
for i in ["Mat", "Inns", "NO", "Runs", "HS", "BF", "100", "50", "0", "4s", "6s"]:
    if "-" in df_batting_avg[i].unique():
        df_batting_avg[i] = df_batting_avg[i].replace("-", None).astype(int)
    else:
        df_batting_avg[i] = df_batting_avg[i].astype(int)
        
for i in ["Ave", "SR"]:
    if "-" in df_batting_avg[i].unique():
        df_batting_avg[i] = df_batting_avg[i].replace("-", None).astype(float)

df_batting_avg.to_csv("Batting_averages.csv", index=False)
df_batting_avg.head()
```

# Bowling averages

```python
url = "http://stats.espncricinfo.com/ci/engine/records/averages/bowling.html?id=12357;type=tournament"
soup = BeautifulSoup(requests.get(url).text, "lxml")
s = soup.select("table", {"class" : "engineTable"})[0].find_all("tbody")[0].find_all("tr", {"class" : "data1"})
columns = soup.select("table", {"class" : "engineTable"})[0].select("tr", {"class" : "head"})[0].get_text().splitlines()[1:]
all_data  = []
for i in s:
    all_data.append(i.get_text().splitlines()[1:])
    df_bowling_avg = pd.DataFrame(all_data, columns = columns)

df_bowling_avg["Country"] = df_bowling_avg.Player.str.extract("\(([^)]+)")
df_bowling_avg["Player"] = df_bowling_avg.Player.str.replace(r"\(.*\)","").str.strip().str.lower()

for i in ["Mat", "Inns", "Mdns", "Runs", "Wkts", "4", "5", "Ct", "St"]:
    if "-" in df_bowling_avg[i].unique():
        df_bowling_avg[i] = df_bowling_avg[i].replace("-", None).astype(int)
    else:
        df_bowling_avg[i] = df_bowling_avg[i].astype(int)
        
for i in ["Overs", "Ave", "Econ", "SR"]:
    if "-" in df_bowling_avg[i].unique():
        df_bowling_avg[i] = df_bowling_avg[i].replace("-", None).astype(float)
    else:
        df_bowling_avg[i] = df_bowling_avg[i].astype(float)
        
        
df_bowling_avg.to_csv("Bowling_averages.csv", index=False)
df_bowling_avg.head()
```

# Players

```python
Teams = ["Australia", "Bangladesh", "England", "India", "New Zealand", "Pakistan",
        "South Africa", "Sri Lanka", "West Indies", "Afghanistan"]

url = "http://www.espncricinfo.com/ci/content/player/"
soup = BeautifulSoup(requests.get(url).text, "lxml")
s = soup.find("div", {"class" : "ciPlayersHomeCtryList"}).select("ul")[0].find_all("li")
country_urls = []
c = Teams.copy()
for i in s:
    try:
        if i.get_text() in Teams:
            country_urls.append(url + i.find("a")['href'])        
            c.remove(i.get_text()) 
    except:
        pass
if c:
    if len(c) == 1:
        if c[0] == "Afghanistan":
            country_urls.append("http://www.espncricinfo.com/ci/content/player/index.html?country=40")
        else:
            raise Exception("There is one country not included to our coutry list, please add it")
    else:
        raise Exception("There is some Countries not included by our scrapper, you should aware it")
        
        
players_links = []
for url in country_urls:
    try:
        soup = BeautifulSoup(requests.get(url).text, "lxml")
        for i in soup.find("table", {"class" : "playersTable"}).select("td", {"class": "divider"}):
            if url != "http://www.espncricinfo.com/ci/content/player/index.html?country=40":
                country = soup.find("li", {"class" : "selectedtab"}).get_text().lower()
            else:
                country = soup.find("span", {"id" : "ciShowCountry"}).get_text().lower()
            players_links.append("http://www.espncricinfo.com/" + country.replace(" ", "") + i.find("a")['href'][3:])
            del country
    except:
        pass
    
    
for url in players_links:
    soup = BeautifulSoup(requests.get(url).text, "lxml")
    
    player_name = soup.find("div", {"class" : "ciPlayernametxt"}).find("h1").find("a")["alt"]
    player_name = player_name.replace(" ", "_").replace("_RSS_Feed", "")
    a = soup.find("div", {"class" : "pnl490M"}).select("table", {"class" : "engineTable"})
    Batting_and_fielding_averages = a[0]
    Bowling_averages = a[1]

    # Batting_and_fielding_averages
    Batting_and_fielding_average_columns = []
    for i in Batting_and_fielding_averages.find("tr", {"class" : "head"}).select("th", {"nowrap" : "nowrap"})[1:]:
        Batting_and_fielding_average_columns.append(i.get_text())
    lst = []
    for i in Batting_and_fielding_averages.find("tbody").find_all("tr", {"class" : "data1"}):
            lst.append(i.get_text().splitlines()[1:])
    df1 = pd.DataFrame(lst, columns= ["Type"] + Batting_and_fielding_average_columns)
#     df1 = df1.add_suffix("_B&F") # Batting_and_fielding

    # Bowling_averages
    Bowling_averages_columns = []
    for i in Bowling_averages.find("tr", {"class" : "head"}).select("th", {"nowrap" : "nowrap"})[1:]:
        Bowling_averages_columns.append(i.get_text())
    lst = []
    for i in Bowling_averages.find("tbody").find_all("tr", {"class" : "data1"}):
            lst.append(i.get_text().splitlines()[1:])
    df2 = pd.DataFrame(lst, columns= ["Type"] + Bowling_averages_columns)
#     df2 = df2.add_suffix("_Bowl") # Bowling

    df = pd.concat([df1, df2.iloc[:, 1:]], axis=1)
    
    country = url.split("/")[3]
    dirs = os.listdir()
    if not country in os.listdir():
        os.makedirs(country)
    
    df.to_csv(f"{country}/{player_name}.csv")
```
