# Scraping scored till 29-may-2019 (before starting WC-2019)

```python
base_url = "http://stats.espncricinfo.com/ci/engine/stats/index.html?class=2;filter=advanced;size=200;spanmax1=29+May+2019;spanval1=span;team=1;team=2;team=25;team=3;team=4;team=40;team=5;team=6;team=7;team=8;template=results;type="
pages_url = "http://stats.espncricinfo.com/ci/engine/stats/index.html?class=2;filter=advanced;page={};size=200;spanmax1=29+May+2019;spanval1=span;team=1;team=2;team=25;team=3;team=4;team=40;team=5;team=6;team=7;team=8;template=results;type="
```

# Batting

```python
url = base_url + "batting"
soup = BeautifulSoup(requests.get(url).text, "lxml")
ran = soup.find("div", {"class" : "pnl650M"}).select("table", {"class" : "engineTable"})[1].\
            find("tr", {"class" : "data2"}).select("td", {"class" : "left"})[1].select("b")
pages = math.ceil(int(ran[2].text) / int(ran[1].text))
urls = [url]

for i in range(2, pages+1):
    urls.append(pages_url.format(i) + "batting")
    
lst = []
for url in urls:
    soup = BeautifulSoup(requests.get(url).text, "lxml")
    s = soup.find("div", {"class" : "pnl650M"}).find_all("table", {"class" : "engineTable"})[2]
    try:
        columns_batting
    except:
        columns_batting = [i for i in s.find("thead").find("tr" , {"class" : "headlinks"}).text.splitlines() if i]
    a = s.find("tbody").select("tr", {"class" : "data1"})
    for i in a:
        record = [z for z in i.text.splitlines() if z]
        lst.append(record)
df_batting = pd.DataFrame(lst, columns=columns_batting)
df_batting.head()
```

# Bowling

```python
url = base_url + "bowling"
soup = BeautifulSoup(requests.get(url).text, "lxml")
ran = soup.find("div", {"class" : "pnl650M"}).select("table", {"class" : "engineTable"})[1].\
            find("tr", {"class" : "data2"}).select("td", {"class" : "left"})[1].select("b")
pages = math.ceil(int(ran[2].text) / int(ran[1].text))
urls = [url]
for i in range(2, pages+1):
    urls.append(pages_url.format(i) + "bowling")
lst = []
for url in urls:
    soup = BeautifulSoup(requests.get(url).text, "lxml")
    s = soup.find("div", {"class" : "pnl650M"}).find_all("table", {"class" : "engineTable"})[2]
    try:
        columns_bowling
    except:
        columns_bowling = [i for i in s.find("thead").find("tr" , {"class" : "headlinks"}).text.splitlines() if i]
    a = s.find("tbody").select("tr", {"class" : "data1"})
    for i in a:
        record = [z for z in i.text.splitlines() if z]
        lst.append(record)
df_bowling = pd.DataFrame(lst, columns=columns_bowling)
df_bowling.head()
```

```python
with open("df_batting.pkl", "rb") as file:
    df_batting = pickle.load(file)
with open("df_bowling.pkl", "rb") as file:
    df_bowling = pickle.load(file)
    
df_batting = df_batting.add_suffix("_bat")
df_batting.rename({"Player_bat" : "Player"}, axis=1, inplace=True)
df_batting["Country"] = df_batting.Player.str.extract("\(([^)]+)")
df_batting["Player"] = df_batting.Player.str.replace(r"\(.*\)","").str.strip().str.lower()
df_batting.head()
```

```python
df_bowling = df_bowling.add_suffix("_bowl")
df_bowling.rename({"Player_bowl" : "Player"}, axis=1, inplace=True)
df_bowling["Country"] = df_bowling.Player.str.extract("\(([^)]+)")
df_bowling["Player"] = df_bowling.Player.str.replace(r"\(.*\)","").str.strip().str.lower()
df_bowling.head()
```

```python
df = df_batting.merge(df_bowling, on="Player", how="outer")
df.drop("Country_x", axis=1, inplace=True)
df.rename({"Country_y" : "Country"}, axis=1, inplace=True)
df.head()
```

# batting ICC ranking

```python
import requests
import os
from bs4 import BeautifulSoup
url = "https://www.icc-cricket.com/rankings/mens/player-rankings/odi/batting"
soup = BeautifulSoup(requests.get(url).text, "lxml")
lst = []
for i in soup.find("tbody", {"class" : "js-list-container"}).select("tr"):
    try:
        new_row = [m for m in [z.strip() for z in i.text.splitlines()] if m]
        if len(new_row) == 3:
            new_row = [int(lst[-1][0])+1] + new_row
        lst.append(new_row)
    except:
        pass
batting_ranks = pd.DataFrame(lst, columns=["Ranking", "Player", "Team", "Points"])
batting_ranks.Player = batting_ranks.Player.str.lower().str.strip()
batting_ranks.head()
```
