```python
import requests
import os
import re
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup

url = "https://www.icc-cricket.com/cricket-world-cup/fixtures"
soup = BeautifulSoup(requests.get(url).text, "lxml")
a = soup.find("div", {"class" : "js-matchlist"})
b = a.select("a", {"class" : "match-block match-block--result t-odi"})
teams1 = []
teams2 = []
stadiums = []
dates = []
winners = []
for i in b:
    c = i.find("div", {"class" : "match-block__meta-container"})
    stadium = c.find("div", {"class" : "match-block__summary"}).\
                                text.strip().split("|")[1].strip()
    date = c.find("time", {"class" : "match-block__date match-block__date--desktop"}).text
    d = i.find("div", {"class" : "match-block__team-container"})
    winner = d.find("div", {"class" : "match-block__result"}).\
                                    text.split("win")[0]
    e = d.find("div", {"class" : "match-block__teams"}).select("div", {"class" : "match-block__team-name"})
    f = []
    for i in e:
        bd = i.text.strip()
        if bd:
            if not bd in f:
                if bd.replace(" ", "").isalpha():
                    f.append(bd)
    teams1.append(f[0])
    teams2.append(f[1])
    stadiums.append(stadium)
    dates.append(date)
    winners.append(winner)
df = pd.DataFrame([teams1, teams2, stadiums, dates, winners]).T
df.columns = ["Team1", "Team2", "Stadium", "Date", "Winner"]
df.to_csv("WC_2019_matches_winners_grounds_and_teams.csv", index=False)
df
```
