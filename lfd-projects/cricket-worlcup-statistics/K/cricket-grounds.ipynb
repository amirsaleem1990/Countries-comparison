```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

with open("cricket-grounds.html", "r") as file:
    file = file.read()
soup = BeautifulSoup(file, "lxml")

d = {}
for i in soup.find("div", {"class" : "PlayersGurusearchBox"}).select("option"):
    a = i.text.split(":")
    if a[0] in d:
        d[a[0]].append(a[1])
    else:
        d[a[0]] = [a[1]]
        
c = 0
for i in d:
    if c < len(d[i]):
        c = len(d[i])

        
for i in d:
    if len(d[i]) < c:
        d[i] = d[i] + [None] * (c - len(d[i]))
        
df = pd.DataFrame(d)
df.to_csv("cricket_grounds_in_all_over_world.csv", index=False)
d
```
