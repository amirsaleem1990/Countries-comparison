```python
url = "https://en.wikipedia.org/wiki/List_of_Pakistani_political_families"
import wikipedia
import os 
import requests
from bs4 import BeautifulSoup
import pandas as pd
r = requests.get(url)
if r.ok:
    html = r.text
    soup = BeautifulSoup(html, 'lxml')
    a = soup.find("div", {"class" : "mw-parser-output"}).findAll("ul")
    names_lst = []
    for e,i in enumerate(a):
        b = i.findAll("li")
        for ii in b:
            c = str(ii.find("a")['href'])
            if not c.startswith("#"):
                if "=edit" not in c:
                    names_lst.append(c[6:])
            if e == 36:
                break
        if e == 36:
                break
names_lst = names_lst[:-1]
```
